### ¿Qué diferencia hay entre la covarianza y la correlación?

La covarianza nos indica si ambas variables varian en la misma dirección (covarianza positiva) o en distinta dirección (covarianza negativa).

La correlación, además de esto, nos indica la fuerza de esa relación. Varía entre -1 y 1

```python
# Cargamos el csv
df = pd.read_csv("../ficheros/diamons_01.csv", index_col= 0)
df.head(2)
```

### Calculamos la covarianza y correlación del dataset

```python
# Hacemos primero el pairplot, para ver la relación visualmente
sns.pairplot(df)
```

```python
# Calculamos la covarianza
df.cov()
```

```python
# Calculamos la correlación
df.corr()
```

Podemos observar tanto visualmente, como con la correlación y la covarianza, que carat (quilates) x (ancho), la y(alto) y la z(profundidad) están fuertemente relacionadas, casi llegando a 1, con una relación positiva. Cuanto mayor es uno, mayor son los otros.

Depth (porcentaje de profundidad) y table (base) no están relacionados con ninguna otra variable ni entre ellos.

Y price, que sería nuestra columna respuesta, está realacionada con todos menos con depth y table. Con carat, x, y y z tiene una fuerte relación positiva. A mayores son, mayor es el precio.
